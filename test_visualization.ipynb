# Visualization of network structure

## Imports

```python
import networkx as nx
import matplotlib.pyplot as plt

import random
```

## Random graph generation + visualization test

```python
NODES = 10
partition = 2

G = nx.Graph()
G.add_nodes_from([i for i in range(NODES)])

edges = []
for i in range(partition):
    for j in range(NODES):
        val = random.randint(0, NODES)
        if j != val:
            edges += [(j, val, random.uniform(1, 2))]

G.add_weighted_edges_from(edges)
plt.subplot(111)
nx.draw(G, with_labels=True)
```
